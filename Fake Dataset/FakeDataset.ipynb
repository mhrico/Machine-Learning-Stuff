# Create a fake dataset with dimensions 1000, 3 which has the following properties:
+ Column 1 must only contain itegers between 1 to 100
+ Coumn 2 must only contain floating point numbers between 1.0000 to 100.0000
+ Column 3 must only contain characters from a-z and A-Z
    

```python
import numpy as np
```

```python
intdata = np.random.randint(0, 100, size=(1000,1))
intdata
```

```python
floatdata = np.random.uniform(1.0, 100.0, size=(1000,1))
floatdata
```

```python
combineddata = np.concatenate((intdata,floatdata),axis=1)
combineddata
```

```python
import string
letters = list(string.ascii_letters)
```

```python
import random
charlist = []
i = 1
while i < 1001:
  charlist = charlist + list(random.choice(letters))
  i += 1
```

```python
chardata = np.array(charlist)
chardata = chardata.reshape(1000,1)
```

```python
soln = np.concatenate((combineddata,chardata),axis=1)
soln
```
